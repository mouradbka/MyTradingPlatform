```python
import sys
import os

#The modules I created to run the different parts of the process. 
sys.path.append('..')
from MyTradingPlatform.DataPipeline.data_acquisition import get_historical_data, get_list_symbols, get_list_coins
from MyTradingPlatform.DataPipeline.feature_engineering import resample, add_cols
from MyTradingPlatform.DataPipeline.models import ARIMA_train, ARIMA_predict
from MyTradingPlatform.DataPipeline.utils import set_binance_key,get_binance_key,set_binance_secret,get_binance_secret

from binance import Client, ThreadedWebsocketManager, ThreadedDepthCacheManager


import pandas as pd
import numpy as np
import math
from datetime import datetime

import mlflow
mlflow.set_tracking_uri('http://localhost:5000')

import mlflow.statsmodels
```

```python
# df = get_historical_data(binance_api_key = '******', \
#                          binance_api_secret = '******',\
#                          interval = '5m')
# df.to_csv('data/btc_all_210422.csv')

df = pd.read_csv('data/btc_all_210422.csv',parse_dates=['OpenTime'] )
df = df.set_index('OpenTime')
```

```python
daily =  add_cols(resample(df))
```

```python
y_train = daily['Close']
y_train.index = pd.DatetimeIndex(y_train.index.values,
                                   freq=y_train.index.inferred_freq)
```

```python
%%time
mlflow.end_run()
name = 'MyTradingPlatform '+str(datetime.now().date())+' ' + str(datetime.now().time() )[:5] #+ 'a'
experiment_id = mlflow.create_experiment(name)
mlflow.set_experiment(name)
# mlflow.start_run()
scores = {}
steps = 1
# for t in [28, 112, 548, len(y_train)]:
for t in [25, 26, 27, 28]:
#     p = 2 
#     d = 1
#     q = 2
    for p in range(6):
            for d in range(6):
                for q in range(6):
                    with mlflow.start_run(nested=True):
                        mlflow.log_param('train_size', t)
                        mlflow.log_param('p', p)
                        mlflow.log_param('d', d)
                        mlflow.log_param('q', q)
                        
                        try:
                            fitmodel = ARIMA_train(y_train.iloc[:-steps], p = p, d = d, q = q, train_size = t)[0]
                            
                            v = y_train.tail(steps)[0]

                            rmse = np.sqrt(fitmodel.mse)
                            try:
                                fore = fitmodel.forecast(steps).apply(lambda x : math.exp(x))[0]
                            except:
                                print(fitmodel.forecast(steps))

                            #MLFlow
                            mlflow.log_metric('rmse' , rmse)
                            mlflow.log_metric('aics' , fitmodel.aic)
                            mlflow.statsmodels.log_model(fitmodel, 'model')
                        #     mlflow.log_metric('training time', time)

                            mlflow.end_run()
                            scores[t] = [rmse, v, fore]
                        except Exception as e:
                            print(e)
                            print('train', t, 'p, d, q', p, d, q)
```
